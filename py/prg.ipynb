```python
##max of three numbers
a = int(input("Enter the first number: "))
b = int(input("Enter the second number: "))
c = int(input("Enter the third number: "))
print(f"The numbers are {a}, {b} and {c}.")
if a>b and a>c:
    print(f"{a} is the largest number.")
elif b>a and b>c:
    print(f"{b} is the largest number.")
else:
    print(f"{c} is the largest number.")
```

```python
##palindrome
num = int(input("Enter the number: "))

temp = num
reverse = 0

while temp>0:
    last_digit = temp % 10
    reverse = (reverse*10) + last_digit
    temp = temp // 10

if num == reverse:
    print(f"{num} is a palindrome number.")
else:
    print(f"{num} is not a palindrome number.")


```

```python
plain = input("Enter the string: ")

temp = plain
reverse = ""

while len(temp)>0:
    last_char = temp[-1]
    reverse = reverse + last_char
    temp = temp[:-1]

if plain == reverse:
    print(f"{plain} is a palindrome string.")
else:
    print(f"{plain} is not a palindrome string.")
```

```python
##Amstrong number
s = 0
num = int(input("Enter the number: "))
temp = num

while temp>0:
    last_digit = temp % 10
    s = s + (last_digit**3)
    temp = temp // 10

if num == s:
    print(f"{num} is an Amstrong number.")
else:
    print(f"{num} is not an Amstrong number.")
```

```python
fact = int(input("Enter the number: "))
temp = fact
f = 1
if fact < 0:
    print("Factorial does not exist for negative numbers.")
elif fact == 0:
    print("The factorial of 0 is 1.")
else: 
    while fact > 0:
        f = f * fact
        fact = fact - 1
    print(f"The factorial of {temp} is {f}.") 
```
