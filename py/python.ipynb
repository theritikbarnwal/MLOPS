```python
#Dynamic typing in python
#python allows you to change the type of a variable during runtime. This is known as dynamic typing. For example, you can assign an integer value to a variable and later assign a string value to the same variable.
var = 10 #initially var is an integer
print(var, type(var))
var = "Hello" #now var is a string
print(var, type(var))
```

```python
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
sum = num1 + num2
diff = num1 - num2
mul = num1 * num2
div = num1 / num2
print(f"The sum of {num1} and {num2} is {sum}.")
print(f"The difference of {num1} and {num2} is {diff}.")
print(f"The product of {num1} and {num2} is {mul}.")
print(f"The division of {num1} and {num2} is {div}.")
```
